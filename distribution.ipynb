```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data_macs = [22, 27, 15, 27, 18, 18, 29, 20, 22, 21, 18, 18, 21, 10, 23, 23, 29, 21, 28, 21, 10, 29, 23, 30,
        18, 8, 21, 13, 12, 27, 20, 27, 18, 29, 28, 21, 26, 30, 29, 26, 31, 19, 15, 28, 27, 28,
        19, 11, 22, 29, 29, 26, 23, 21, 19, 21, 26, 31, 27, 22, 14, 27, 23, 18, 22]
data_sorted_macs = sorted(data_macs)
print(f"Attending Students: {len(data_macs)}")

data_min, data_max = min(data_sorted_macs), max(data_sorted_macs)
# Generate bins from min to max with a step of 1
bins = np.arange(data_min - 0.5, data_max + 1.5, 1)
mean_value = np.mean(data_sorted_macs)
median_value = np.median(data_sorted_macs)

plt.figure(figsize=(10, 6))
# Plot histogram with manually specified bins
plt.hist(data_sorted_macs, bins=bins, alpha=1, color='blue', rwidth= 0.5)
plt.title('Macro Midterm')
plt.xlabel('Valutazione')
plt.ylabel('Frequenza')
plt.grid(True)

# Set x-axis ticks to label each unique value
plt.xticks(ticks=np.arange(data_min, data_max + 1), labels=np.arange(data_min, data_max + 1))
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=1)
plt.axvline(median_value, color='green', linestyle='dashed', linewidth=1)

# Adding legend to differentiate mean and median
plt.legend({'Media': mean_value, 'Mediana': median_value})

# Set x-axis ticks to label each unique value
plt.xticks(ticks=np.arange(data_min, data_max + 1), labels=np.arange(data_min, data_max + 1))

plt.show()
plt.show()
```
